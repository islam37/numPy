```python



# Create a 2D array
arr = np.array([[1, 2, 3],
                [4, 5, 6]])

print("Array:")
print(arr)

# Shape of the array
print("Shape:", arr.shape)

# Number of dimensions
print("Dimensions:", arr.ndim)

# Basic operations
print("Sum:", np.sum(arr))
print("Mean:", np.mean(arr))
print("Max:", np.max(arr))
```
